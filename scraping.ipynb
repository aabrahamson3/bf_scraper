```python
def get_recipes_html(r):
    soup = BeautifulSoup(r.content, 'html.parser')
    recipes = soup.select('.recipetitle')
    return recipes
```

```python
def add_to_recipe_dict(recipes, recipe_dict):
    for recipe in recipes:
        recipe_dict[recipe.get_text()] = recipe['href']
    return recipe_dict
```

```python
def navigate_tables(table, brewpart):
    easy_brewparts = ["fermentables", "hops", "mashsteps", "others"]
    if brewpart in easy_brewparts:
        scrape_easy_brewparts(table, brewpart)
    elif brewpart == "yeasts":
        scrape_yeast(table, brewpart)
    elif brewpart == 'primingmethod':
        scrape_priming(table, brewpart)
    elif brewpart == 'water':
        scrape_water(table, brewpart)

        
def scrape_notes(table, brewpart):
    table_body = table.find('a', name="notes")
    notes = table_body.find('div', class_="ui message")
    out_dict['notes'] = notes.text

def scrape_water(table, brewpart):
    water_list = []
    table_body = table.find('tbody')
    elements = table_body.find_all('td')
    for element in elements:
        water_list.append(element.text.strip())
    water_list = water_list[:-1]
    out_dict[brewpart] = water_list
           
        
def scrape_priming(table, brewpart):
    text = table.find('td').text.strip()
    out_dict[brewpart] = text
    
        
def scrape_yeast(table, brewpart):        
    yeast_list = []
    strain = table.find('span').text
    yeast_list.append(strain)
    table_body = table.find('tbody')
    elements = table_body.find_all('td')
    for i in range(1,len(elements)):
        if i % 2 == 0:
            text = elements[i].text.strip()
            text = text.replace('\n','')
            text = text.replace('\t','')
            text = text.replace(u'\xa0', u' ')

            yeast_list.append(text)
    out_dict[brewpart] = yeast_list
        

def scrape_easy_brewparts(table, brewpart):
    table_body = table.find('tbody')
    rows = table_body.find_all('tr')
    brewpart_list = []
    for row in rows:
        cols = row.find_all('td')
        cols = strip_elements(cols)
        brewpart_list.append(cols)
    out_dict[brewpart] = brewpart_list

def strip_elements(row):
    out_list = []
    for i in range(len(row)):
        if i != 1:
            out_list.append(row[i].text.strip())
        else:
            if row[i].find('a') is None:
                out_list.append('None')
            else:
                out_list.append(row[i].find('a').text)
            
    return out_list

def split_ingredient_row(row):
    try:
        split = row[1].split(row[0][:2])[0]
        row[1] = split
        return row
    except:
        return row
    
```
